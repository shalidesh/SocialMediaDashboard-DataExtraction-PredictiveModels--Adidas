```python
import pandas as pd

# Assuming df is your DataFrame and 'Uploaded Date' is in datetime format
df['Title Length'] = df['Title'].apply(len)
df['Uploaded Day of Week'] = df['Uploaded Date'].dt.day_name()
df['Uploaded Time of Day'] = df['Uploaded Date'].dt.hour
df['Uploaded Month'] = df['Uploaded Date'].dt.month

df.head()
```

```python
# Remove columns using drop() method
new_df = df.drop(columns=['PostID','Post URL','Uploaded Date'], axis=1)
new_df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

nltk.download('stopwords')
```

```python
import nltk
nltk.download('punkt')
```

```python

# Function to preprocess text
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove punctuation
    text = re.sub('[^a-z]', ' ', text)

    # Remove stopwords and stem the words
    stemmer = PorterStemmer()
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(text)
    words = [stemmer.stem(word) for word in words if word not in stop_words]

    # Join the words back into a single string
    text = ' '.join(words)

    return text

# Apply the function to the 'Title' column
new_df['Processed Title'] = new_df['Title'].apply(preprocess_text)
```

```python
new_df1 = new_df.drop(columns=['Title'], axis=1)
new_df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()
```

```python
# Fit and transform on your training data
new_df1['Uploaded Day of Week'] = le.fit_transform(new_df1['Uploaded Day of Week'])
new_df1.head()
```

```python
new_df1.shape
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming numerical_cols is a list of the names of all numerical columns
numerical_cols = ['Title Length', 'Uploaded Time of Day','Uploaded Month','Likes', 'Comments','Uploaded Day of Week']

scaler = StandardScaler()
new_df1[numerical_cols] = scaler.fit_transform(new_df1[numerical_cols])
new_df1.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the 'Processed Title' column
X = vectorizer.fit_transform(new_df1['Processed Title'])

# Now, X is a matrix where each row corresponds to a document (title)
# and each column corresponds to a word. The value in each cell is the
# TF-IDF score of that word in that document.
```

```python
X
```

```python
# Convert X to a DataFrame
tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Concatenate the original DataFrame with the TF-IDF DataFrame
df = pd.concat([new_df1, tfidf_df], axis=1)
```

```python
df.head()
```

```python
df1 = df.drop(columns=['Processed Title'], axis=1)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define your features and target variable
X = df1.drop('Shares', axis=1).values
y = df1['Shares'].values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set and evaluate the model
y_pred = model.predict(X_test)
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Assuming y_test is your test target values and y_pred are the predicted values from your model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

```python
import numpy as np

def predict_virality(title, uploaded_date, likes, comments):

  # Preprocess the title
  processed_title = preprocess_text(title)

  # Extract features from the uploaded date
  uploaded_day_of_week = uploaded_date.day_name()
  uploaded_time_of_day = uploaded_date.hour
  uploaded_month = uploaded_date.month


  # Convert the preprocessed title to TF-IDF features
  tfidf_features = vectorizer.transform([processed_title])

  # Convert the day of week to one-hot encoded features
  # To transform single record
  day_of_week_features = le.transform([uploaded_day_of_week])
  # Normalize numerical features
  numerical_features = scaler.transform([[likes, comments,len(title),day_of_week_features,uploaded_time_of_day,uploaded_month,]])
  # Concatenate all features into a single feature vector

  features = np.concatenate([numerical_features,tfidf_features.toarray()], axis=1)
  # Make a prediction with the model
  shares_pred = model.predict(features)

  return shares_pred[0]
```

```python
# Define the raw data
title = "Check out our new product line!"
uploaded_date = pd.to_datetime('2023-10-06 12:00:00')
likes = 5
comments = 5
```

```python

result=predict_virality(title, uploaded_date, likes, comments)
print(f' Predictable Share Count on Social Media : {result}')
```
